```python
from imblearn.over_sampling import BorderlineSMOTE
Borderline = BorderlineSMOTE(random_state=42)

x_train_res, y_train_res = Borderline.fit_resample(x_train, y_train)
```

```python
import torch
from tab_transformer_pytorch import FTTransformer

# number of categorical columns
num_cat = x_train_res.shape[1]

model = FTTransformer(
    categories = tuple([2] * num_cat),  # each feature has {0,1}
    num_continuous = 0,
    dim = 15,
    dim_out = 1,
    depth = 4,
    heads = 8
)

# Convert dataframe to tensor
x_categ = torch.tensor(
    x_train_res.values, dtype=torch.long
)

x_numer = None

pred = model(x_categ, x_numer)
```

```python
y_train_tensor = torch.tensor(
    y_train_res.values, dtype=torch.float32
)
import torch.nn as nn
import torch.optim as optim

criterion = nn.BCEWithLogitsLoss()
optimizer = optim.Adam(model.parameters(), lr=1e-3)
model.train()

for epoch in range(20):
    optimizer.zero_grad()

    logits = model(x_categ, None)
    loss = criterion(logits.squeeze(), y_train_tensor)

    loss.backward()
    optimizer.step()

    print(f"Epoch {epoch+1}, Loss: {loss.item():.4f}")
model.eval()
with torch.no_grad():
    logits = model(x_categ, None)
    probs = torch.sigmoid(logits)
    y_pred = (probs >= 0.5).int()
```

```python
from sklearn.metrics import classification_report,accuracy_score

y_true = y_train_tensor.cpu().numpy()
y_pred_np = y_pred.squeeze().cpu().numpy()
print(classification_report(y_true, y_pred_np))
print(f"Accuracy: {accuracy_score(y_true, y_pred_np):.2f}")

```
