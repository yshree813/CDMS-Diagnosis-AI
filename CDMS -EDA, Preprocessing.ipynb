```python
data['Gender']=data['Gender'].replace(2,0)
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer(
    transformers=[
        ('cat', OneHotEncoder(
            handle_unknown='ignore',
            sparse_output=False
        ), ['Varicella',
    'Mono or polysymptomatic',
    'Oligoclonal bands',
    'initial EDSS','initial symptom',
    ' final EDSS'])
    ],
    remainder='passthrough'
)

data_encoded = ct.fit_transform(data)

```

```python
feature_names = ct.get_feature_names_out()
```

```python
data_encoded['remainder__group']=data_encoded['remainder__group'].replace(2,0)
```

```python
from sklearn.feature_selection import mutual_info_classif

mi = mutual_info_classif(data_encoded.drop(columns=['remainder__group']), data_encoded['remainder__group'], random_state=42)

mi_df = pd.DataFrame({
    'feature': data_encoded.drop(columns=['remainder__group']).columns,
    'MI_score': mi
}).sort_values(by='MI_score', ascending=False)
```

```python
from sklearn.model_selection import train_test_split
x=data_encoded.drop(columns=['remainder__group'])
y=data_encoded['remainder__group']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
import pandas as pd

ct = ColumnTransformer(
    transformers=[
        ('scale_age', StandardScaler(), ['remainder__Age (y)'])
    ],
    remainder='passthrough'
)

x_scaled_train = ct.fit_transform(x_train)

x_scaled_train = pd.DataFrame(
    x_scaled_train,
    columns=ct.get_feature_names_out(),
    index=x_train.index
)


x_scaled_test = ct.transform(x_test)

x_scaled_test = pd.DataFrame(
    x_scaled_test,
    columns=ct.get_feature_names_out(),
    index=x_test.index
)
```

```python
from sklearn.ensemble import RandomForestClassifier

model = RandomForestClassifier(random_state=42, n_jobs=-1)
model.fit(x_scaled_train, y_train)
y_pred = model.predict(x_scaled_test)
y_prob = model.predict_proba(x_scaled_test)[:,1]
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, average_precision_score,accuracy_score

print("Accuracy:", accuracy_score(y_test, y_pred))

print(classification_report(y_test, y_pred))
print("Confusion matrix:\n", confusion_matrix(y_test, y_pred))
print("ROC AUC:", roc_auc_score(y_test, y_prob))
print("PR AUC (average precision):", average_precision_score(y_test, y_prob))
```
