```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
data=pd.read_csv('/content/data_final preprocessed_cdms.csv')
data
```

```python
x=data.drop('remainder__group',axis=1)
y=data['remainder__group']
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=42)
```

```python
# !pip install --upgrade imbalanced-learn
from imblearn.over_sampling import BorderlineSMOTE
Borderline = BorderlineSMOTE(random_state=42)

x_train_res, y_train_res = Borderline.fit_resample(x_train, y_train)
```

```python
import torch
x=torch.tensor(x_train_res.values, dtype=torch.float32)
y=torch.tensor(y_train_res.values,dtype=torch.long)
```

```python
import torch.nn as nn
from models.model import RowColTransformer

print("Raw X shape:", x.shape)

# Prepare input ONCE
x = x.unsqueeze(-1)
print("After unsqueeze:", x.shape)

# Projection: 1 → 16
proj = nn.Linear(1, 12)

# SAINT backbone (small)
opt = RowColTransformer(
    num_tokens=0,
    dim=12,
    nfeats=x.shape[1],
    depth=2,
    heads=8,
    dim_head=8,
    attn_dropout=0.1,
    ff_dropout=0.1,
    style="col"
)

# ✅ FIXED classifier head
clf_head = nn.Linear(12, 2)

criterion = nn.CrossEntropyLoss()

optimizer = torch.optim.Adam(
    list(proj.parameters()) +
    list(opt.parameters()) +
    list(clf_head.parameters()),
    lr=1e-3
)

# Training loop
for epoch in range(20):
    optimizer.zero_grad()

    x_proj = proj(x)  # (N, F, 16)
    x_proj = x_proj.view(x_proj.size(0), x_proj.size(1), -1)

    features = opt(x_proj)          # (N, F, 16)
    features = features.mean(dim=1) # (N, 16)

    logits = clf_head(features)     # (N, 2)

    loss = criterion(logits, y)
    loss.backward()
    optimizer.step()

    print(f"Epoch {epoch+1:02d} | Loss: {loss.item():.4f}")
```

```python
from sklearn.metrics import accuracy_score, classification_report
x_test=torch.tensor(x_test.values, dtype=torch.float32)
y_test=torch.tensor(y_test.values,dtype=torch.long)

# make sure test input has correct shape
x_test = x_test.unsqueeze(-1)   # (N_test, F, 1)
y_test_t = y_test               # already tensor

# eval mode
proj.eval()
opt.eval()
clf_head.eval()

with torch.no_grad():
    x_test_proj = proj(x_test)
    features = opt(x_test_proj)
    features = features.mean(dim=1)
    logits = clf_head(features)
    preds = torch.argmax(logits, dim=1)

print("Accuracy:", accuracy_score(y_test_t.numpy(), preds.numpy()))
print(classification_report(y_test_t.numpy(), preds.numpy()))
```
